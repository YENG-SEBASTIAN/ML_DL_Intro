```python
import numpy as np

a = np.array([1, 2, 3, 4, 5])
b = np.array([[1, 2, 3],
              [4, 5, 6]])

print(a)
print(b)
```

```python
print("Shape of a:", a.shape)
print("Shape of b:", b.shape)

print("number of dimensions of a:", a.ndim)
print("number of dimensions of b:", b.ndim)
```

```python
X = np.array([1, 2, 3, 4, 5])

print("X shape before:", X.shape)

X = X.reshape(-1, 1)
print("X shape after reshape:", X.shape)
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 30, 40])

print(x + y)
print(x * y)
print(x ** 2)
```

```python
data = np.array([12, 15, 20, 22, 25, 30])

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Std Dev:", np.std(data))
```

```python
X = np.array([100, 120, 140, 160, 180]).reshape(-1, 1)
y = np.array([200, 240, 280, 320, 360])

print(X)
print(y)
```

```python
X = np.array([5,10,15,20,25,30,35,40,45,50])

print("elements:", X)

new_shape = X.reshape(10, 1)
print("reshaped elements:", new_shape)

print("Mean:", np.mean(new_shape))
print("Median:", np.median(new_shape))
print("Std Dev:", np.std(new_shape))
print("Max:", np.max(new_shape))
print("Min:", np.min(new_shape))
print("square:", np.square(new_shape))
```
