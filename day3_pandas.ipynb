```python
import pandas as pd

data = {
    "experience_years": [1, 2, 3, None, 5, 6],
    "salary": [30000, 35000, 40000, 45000, None, 55000]
}

df = pd.DataFrame(data)
df

df.isna()
df.dropna()
new_data = df.fillna(df.mean(numeric_only=True))
new_data

X = new_data["experience_years"].to_numpy()
y = new_data["salary"].to_numpy()

average_salary = y.mean()
print(f"Average Salary: {average_salary}")

new_data["salary"] > average_salary
```
